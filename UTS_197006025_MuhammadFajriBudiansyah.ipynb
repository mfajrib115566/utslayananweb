```python
df_10= df.head(10)

fig = plt.figure(figsize=(20,3))
ax = fig.add_axes([0,0,1,1])
ax.set_xlabel('Player Name')
ax.set_ylabel('Goals')

ax.bar(df_10['player_name'],df_10['goals'])
plt.title('Pencetak Gol')
plt.show()
```

Perintah diatas merupakan membuat data dalam bentuk histogram dimana saya ingin mengetahui data dari 10 pemain teratas dan berapa gol yang telah dicetak pemain tersebut.

```python
player_name = ('Harry', 'Salah', 'Bruno', 'Soo', 'Bamford')
red_cards = (0, 0, 0, 0, 0)
df = pd.DataFrame({
'Player': player_name,
'Card': red_cards,
})
df.sort_values(by='Card', inplace=True)
x_coords = np.arange(len(df))
plt.bar(x_coords, df['Card'], tick_label=df['Player'])
plt.xticks(rotation=90)
plt.ylabel('Red Cards')
plt.title('Player Name')
plt.show()
```

```python
player_name = ('Harry', 'Salah', 'Bruno', 'Soo', 'Bamford')
yellow_cards = (1, 0, 6, 0, 3)
df = pd.DataFrame({
'Player': player_name,
'Card': yellow_cards,
})
df.sort_values(by='Card', inplace=True)
x_coords = np.arange(len(df))
plt.bar(x_coords, df['Card'], tick_label=df['Player'])
plt.xticks(rotation=90)
plt.ylabel('Yellow Cards')
plt.title('Player Name')
plt.show()
```
